```python
'''Logistic Regression''
    predicts the probability of poor care
        Denote dependent variable "Poor care" by y
        P(y=1)
    Then P(y=0)=1-P(y=1)
    INdependent variables x1,x2,x3,......xk
    uses the logistic regression response function:
        P(y=1)=1/(1+e^-(beta0+beta1x1+beta2x2+......betakxk))
        non linear transformation of linear regression equation to produce number between 0 and 1
    positive values are predict of class 1
    negative values are predict of class 0
    The coefficients are selected to:
        predict a high probability for the poor care cases
        predict a low probability for the good care cases
    we can instead talk about the odds(like in gambling)
        odds=P(y=1)/P(y=0)
    odds>1, if y=1 more likely
    odds<1, if y=0 more likely
    Turns out that:
        odds=e^(beta0+beta1x1+beta2x2+.......)
        log(odds)=beta0+beta1x1+beta2x2+.....
    This is the logit and looks like linear regression
    The bigger the logit is, the bigger P(y=1)
    
    
        
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import random
import statistics
import statsmodels.api as sm
import sklearn
import scikitplot as skplt
quality=pd.read_csv('./quality.csv')
quality.head()
quality.info()#structure of the quality dat
quality.PoorCare.value_counts()#counting the number of poor care and good care
#logistic regression is method of classification, a standard baseline method is to just predict the most frequent outcome for all observations
#As we can see, the quality of good care has 98 out of 131 observation, so the accuracy is 74.8%. we will beat that by using the method of logistic regression
# we will see how we can split th dataset into two parts, training set and testing set
# quatrain=quality[]
# random.seed(88)#we have set the seed in  order to randomly split the data
indices=range(quality.shape[0])#this would be the range og the rows of the data set
numtrain=int(0.75*quality.shape[0])#how much do we wish to split the data, in our case we are splitting in 75-25 ratio
np.random.shuffle(indices)#random shuffle the range
quatrainind=indices[:numtrain]#train would be the 75% of data randomy shuffeled
quatestind=indices[numtrain:]#test would be the remaining
quatrain=quality.iloc[quatrainind]#converting it to the data frame train
quatest=quality.iloc[quatestind]#converting it to the data frame test
# we have to make sure that spliting done in a way that in both sets, percentahe of good care are almost same
len(quatrain)
#Now lets build the logistic regression model using OfficeVisits and Narcotics are independent variable
# cols=['OfficeVisits','Narcotics']#making a list of the variables or say independent variables
# X=quatrain[cols]#setting a independent variables in one variable
# Y=quatrain["PoorCare"]#dependent variable
result=sm.formula.glm("PoorCare ~ OfficeVisits+Narcotics",family=sm.families.Binomial(),data=quatrain).fit()#building a logistic regression model
result.summary()
#lets have a prediction on test set
res=result.predict(quatrain)
#Thresholding
#The outcome of a logistic regression model is a probability
#often we want to make a binary prediction
#we can do that by threshold value t
#if P(PoorCare=1)>=t, predct poor quality
#if P(PoorCare=0)>=t, predct good quality
#often select based on which error are better
#if t is large, predict poor quality rarely
    #more error where we say good care, but it is actually poor care
    #detect patients who are recieving the worst care
#if t is small, predict good care rarely
    #more errors wee say poor care, but actually good care
    #detects all patients who might be recieving poor care
#with no preferences between errors, select t =0.5
    #predicts more likely outcome

#CONFUSION MATRIX
    #compare actual outcomes to predicted outcomes using a confusion matrix
    
    #Sensitivity=TP/(TP+FN)
    #specificity=TN/(TN+FP)
actual=np.array(quatrain.PoorCare)
pred=np.array(res)
pd.crosstab(actual,pred>0.5)#set  a threshold of 50%
#ROC(Reciever operator characteristic) Curve which can help to decide which threshold is the best
    #True positive rate-proportion of poor care caught
    #false positive-proportion of good caare labelled as poor care
    #lablled as (x,y)-->[0=<(X,y)<=1]
        #x is labelled for False postitve rate
        #y is labelled for true positive rate
    #The higher the threshold, the lower the sensitivity and higher the specificity
    # the lower the threshold, the higher the sensitivity and lower the specificity
    #So which threshold you should pick?
        #If you are more concerened with having a high specificity or low false positive rate, pick threshold that maximize true positive rate
        #If you are more concerened with having a high sensitivity, pick threshold that maximize false positive rate

auc=np.trapz(pred,actual)
# plt.plot(actual,pred)
# plt.show()
skplt.metrics.plot_roc_curve(x,y)
```

```python
import pandas as pd
import numpy as np
import os
import statsmodels.api as sm
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import  train_test_split#this time we are using training and testing split by sklearn
import statistics
import random
fram=pd.read_csv('./framingham.csv')
fram.info()
train,test=train_test_split(fram,train_size=0.75,random_state=88)
framtrain=pd.DataFrame(train,columns=fram.columns)
framtest=pd.DataFrame(test,columns=fram.columns)
logistic=sm.formula.glm("TenYearCHD ~ male+age+education+currentSmoker+cigsPerDay+BPMeds+prevalentStroke+prevalentHyp+diabetes+totChol+sysBP+diaBP+BMI+heartRate+glucose",family=sm.families.Binomial(),data=framtrain).fit()
logistic.summary2()
predtest=logistic.predict(framtest)
pd.crosstab(framtest.TenYearCHD,predtest>0.5)
pd.crosstab(fram.TenYearCHD,predtest>0.5)
```
