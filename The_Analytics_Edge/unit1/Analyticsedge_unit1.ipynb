```python
'''Adding Variables'''
import pandas as pd
import numpy as np
import os
import statistics
import matplotlib.pyplot as plt
data=pd.read_csv('./USDA.csv')
data.Sodium[50]>statistics.mean(data.Sodium.dropna())#checking if the 50th food of sodium is greater than mean value of sodium observation or not\
data['HighSod']=pd.to_numeric(data.Sodium>statistics.mean(data.Sodium.dropna()))#adding a new column to data frame and converting it to numeric
data['HighPro']=pd.to_numeric(data.Protein>statistics.mean(data.Protein.dropna()))
data['HighFat']=pd.to_numeric(data.TotalFat>statistics.mean(data.TotalFat.dropna()))
data['HighCarbs']=pd.to_numeric(data.Carbohydrate>statistics.mean(data.Carbohydrate.dropna()))
datanew=data[data.HighSod==True]
data.HighSod.value_counts()#making a table of counting number of true and false in HighSod variable
pd.crosstab(data.HighSod.dropna(),data.HighCarbs.dropna())#comparing two tables
a=data.pivot_table(index='Iron',columns='HighCarbs',aggfunc=np.mean)
a.mean
```
